```python
import numpy as np
import matplotlib.pyplot as plt
```

# Question 1

```python
def density(r, a=10):
    return (1/r)*(2*r+3*a)/(r+a)
```

```python
r = np.linspace(0, 50,50)
d=[]
for i in r:
    d.append(density(i))

plt.plot(r,d)

plt.grid('on')
#plt.title('Comparing Density to Radius')
plt.xlabel('Radius (kpc)')
plt.ylabel("Log(Density)")
plt.yscale('log')
plt.savefig("density.png")
```

```python
def A(r, a=10):
    return r/a

def v(r, G=6.67*10**-11, a=10, p_0=3.5*10**-21):
    return np.sqrt(4*G*p_0*a^2(np.log(1+A(r))/A(r)-1/(1+A(r))))

def v_kep(r,M=1.989*10**30*10**11,G=6.67*10**-11):
    return np.sqrt(G*M/r)
```

```python
v_c = []
for i in r:
    v_c.append(A(i))

v_k = []
r_2 = np.linspace(15, 50, 50)
for i in r_2:
    v_k.append(v_kep(i))

plt.plot(r, v_c)

plt.xlabel('Radius (kpc)')
plt.ylabel("V_circ (km/s)")
plt.grid('on')
plt.savefig("v_circ.png")
```

```python
plt.plot(r_2, v_k)
plt.xlabel('Radius (kpc)')
plt.ylabel("V_keplarian (km/s)")
plt.grid('on')
plt.savefig("v_kep.png")
```
